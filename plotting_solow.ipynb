```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("dark_background")
import numpy as np
```

```python
alpha = 1/3
```

```python
k = np.arange(0, 100, 0.1)
```

```python
output_per_capita =  k**alpha
```

```python
capital_acc = 0.45*output_per_capita
```

```python
maxi = np.array(output_per_capita).max()+1
maxi
```

```python
np.intersect1d(np.arange(0, 3, 0.01), capital_acc)
```

```python
k = np.arange(0, 100, 0.1)
```

```python
capital_acc = 0.4*output_per_capita
```

```python
fig, ax = plt.subplots(1)

ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))
# Hide the top and right spines.
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Draw arrows (as black triangles: ">k"/"^k") at the end of the axes.  In each
# case, one of the coordinates (0) is a data coordinate (i.e., y = 0 or x = 0,
# respectively) and the other one (1) is an axes coordinate (i.e., at the very
# right/top of the axes).  Also, disable clipping (clip_on=False) as the marker
# actually spills out of the axes.
ax.plot(1, 0, ">w", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^w", transform=ax.get_xaxis_transform(), clip_on=False)

ax.plot(k, output_per_capita, label="y=$k^Î±$")
ax.plot(k, capital_acc, label="sy")
ax.plot(np.arange(0, 3, 0.01)*30, np.arange(0, 3, 0.01), label="$(n+\delta)k$")
plt.xticks([])
plt.yticks([])
plt.ylim(top=maxi)
plt.xlabel("k", loc="right", weight='bold')
plt.title("The Solow Model")
plt.legend(loc="upper center")
plt.axvline(x=41.5, ymin=0.04, ymax=0.64, color='r', linestyle='--')
plt.axhline(y=1.39, xmin=0.05, xmax=0.429, color='r', linestyle='--')
plt.axhline(y=3.45, xmin=0.05, xmax=0.429, color='r', linestyle='--')
plt.text(41.5, -0.23,"$k^*$")
plt.text(-4, 3.45,"$y^*$")
plt.text(-6, 1.39,"$sy^*$")
plt.arrow(30, 1.229, dx=0.01, dy=0, shape='full', color="yellow", lw=0, length_includes_head=True, head_width=0.16, head_length=2.1)
plt.arrow(25, 1.15, dx=0.01, dy=0, shape='full', color="yellow", lw=0, length_includes_head=True, head_width=0.16, head_length=2.1)
plt.arrow(55, 1.53, dx=-0.01, dy=0, shape='full', color="yellow", lw=0, length_includes_head=True, head_width=0.16, head_length=2.1)
plt.arrow(60, 1.57, dx=-0.01, dy=0, shape='full', color="yellow", lw=0, length_includes_head=True, head_width=0.16, head_length=2.1)
plt.arrow(38, 1.33, dx=0.01, dy=0, shape='full', color="yellow", lw=0, length_includes_head=True, head_width=0.16, head_length=2.1)
plt.arrow(45, 1.43, dx=-0.01, dy=0, shape='full', color="yellow", lw=0, length_includes_head=True, head_width=0.16, head_length=2.1)
plt.show()
```
