```python
fig, ax = plt.subplots()

# Iteration Ã¼ber eindeutige LÃ¤nder in der Spalte "LOCATION"
for country, group in df.groupby('LOCATION'):
    if country in ["DEU", "CHN", "JPN", "USA", "CHE"]:
        group.plot(y='Value', ax=ax, label=country, grid=False)
plt.title("Economic Growth")
plt.legend(loc="best")
plt.ylabel("GDP per capita in $")
plt.xlabel("Year")
plt.tight_layout()
ax.yaxis.set(ticks=np.arange(0, 110000, 10000))
ax.set_yticklabels(['{:,}'.format(int(x)) for x in ax.get_yticks().tolist()])
```

```python
alpha = 1/4
```

```python
k = np.arange(0, 100, 0.1)
```

```python
output_per_capita =  k**alpha
```

```python
capital_acc = 0.4*output_per_capita
```

```python
fig, ax = plt.subplots(1, figsize=(16, 8))
ax.plot(k, output_per_capita)
ax.plot(k, capital_acc)
ax.plot(np.arange(0, 5, 0.1)*30, np.arange(0, 5, 0.1))
plt.xticks([])
plt.yticks([])
plt.xlabel("k = K/L")
plt.show()
```
